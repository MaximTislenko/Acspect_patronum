```
def issled (text, per_list):
  pos_lab_start = text.find(per_list[0], text.find(per_list[0])+1)
  pos_lab_end = text.find(per_list[1], text.find(per_list[1])+1)
  text_lab = text[pos_lab_start + 41  : pos_lab_end - 5 ]
  pos_ins_start = text.find(per_list[1], text.find(per_list[1])+1)
  pos_ins_end = text.find(per_list[2], text.find(per_list[2])+1)
  text_ins = text[pos_ins_start : pos_ins_end]
  pos_in_start = text.find(per_list[2], text.find(per_list[2])+1)
  pos_in_end = text.find(per_list[3], text.find(per_list[3])+1)
  text_in = text[pos_in_start : pos_in_end]
  # print(text_lab)
  return text_lab, text_ins, text_in

per_list = ['Лабораторные диагностические исследования', 'Инструментальные диагностические исследования', 'Иные диагностические исследования', '3. Ле']

text_lab, text_ins, text_in  = issled (text, per_list)
```

```
# Подгрузка библиотек
import requests as req
import pandas as pd
from datetime import datetime

# Парсинг + Запрос API
link = 'https://apicr.minzdrav.gov.ru/api.ashx?op=GetJsonClinrecs&ssid=undefined'
response = req.get(link).json()

# Модификация и формирование таблицы
base = {'id': [],
        'status': [],
        'db_id':[],
        'code': [],
        'version': [],
        'ver': 1,
        'age': [],
        'name': [],
        'Mkb': [],
        'ass': [],
        'publish_date': [],
        'NPC_approved': [],
        'apply_status': [],
        'prev_cr_id': []}
df = pd.DataFrame(base)
for i in range(len(response)):
  text = response[i].get('publish_date')
  text_01 = int(text[6:-2])/1000 + 10800
  date = datetime.fromtimestamp(text_01).strftime('%d-%m-%Y')
  up = {'publish_date': date}
  response[i].update(up)
  df = pd.concat([df, pd.DataFrame.from_records(response[i])] ,ignore_index=True)
  up = {}

#Сохранение результата
klin_rec = df.copy()
klin_rec.to_csv('/content/sample_data/klin_rec.csv', header=True)

# Взрослые
# klin_rec_adults = klin_rec[((klin_rec['age'] == 1) | (klin_rec['age'] == 3))]
# klin_rec_adults.to_csv('/content/sample_data/klin_rec_adults.csv', header=True)
# klin_rec_adults.to_excel('/content/sample_data/klin_rec_adults.xlsx', header=True)

# Дети
# klin_rec_сhildren = klin_rec[((klin_rec['age'] == 2) | (klin_rec['age'] == 3))]
# klin_rec_сhildren.to_csv('/content/sample_data/klin_rec_сhildren.csv', header=True)
# klin_rec_сhildren.to_excel('/content/sample_data/klin_rec_сhildren.xlsx', header=True)
```
