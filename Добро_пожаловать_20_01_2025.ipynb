```
L = -1.7862
M = 16.9392
S = 0.11070
BMI = 30
```

```
z_ind = (((BMI/M)**L)-1)/(S*L)
print(z_ind)
```

```
SD3_poz = M*(1+L*S*3)**(1/L)
print(SD3_poz)
```

```
SD2_poz = M*(1+L*S*2)**(1/L)
print(SD2_poz)
```

```
sd23_pos = SD3_poz - SD2_poz
print(sd23_pos)
```

```
z_ind_real = 3 + ((BMI - SD3_poz)/sd23_pos)
print(z_ind_real)
```

```
df = pd.DataFrame({'age':(132, 162, 108, 3, 4),
                   'L' : (-1.7862, -1.3529, -1.6318, 0.5512, 0.8568),
                   'M' : (16.9392, 20.4951, 16.0490, 4.5618, 0.1234),
                   'S' : (0.11070, 0.12579, 0.10038, 0.4567, 0.8912)})
```

```
# @title ФУНКЦИЯ
def calculation (age, BMI):
  L = df[df['age'] == age].iat[0, 1]
  M = df[df['age'] == age].iat[0, 2]
  S = df[df['age'] == age].iat[0, 3]
  z_ind = (((BMI/M)**L)-1)/(S*L)
  if z_ind > 3:
    z_ind = 3 + (BMI - M*(1+L*S*3)**(1/L))/(M*(1+L*S*3)**(1/L) - M*(1+L*S*2)**(1/L))

  if z_ind < -3:
    z_ind = -3 + ((BMI - (M*(1+L*S*-3)**(1/L)))/((M*(1+L*S*-2)**(1/L)) - (M*(1+L*S*-3)**(1/L))))

  print(z_ind)
```

```
calculation (108, 19)
```

```
calculation(132, 30)
```

```
calculation(162, 14)
```
