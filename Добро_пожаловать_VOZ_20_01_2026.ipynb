# Если >3

```
L = -1.7862
M = 16.9392
S = 0.11070
BMI = 30
```

```
z_ind = (((BMI/M)**L)-1)/(S*L)
print(z_ind)
```

```
SD3_poz = M*(1+L*S*3)**(1/L)
print(SD3_poz)
```

```
SD2_poz = M*(1+L*S*2)**(1/L)
print(SD2_poz)
```

```
sd23_pos = SD3_poz - SD2_poz
print(sd23_pos)
```

```
z_ind_real = 3 + ((BMI - SD3_poz)/sd23_pos)
print(z_ind_real)
```

##Итоговая формула > 3

```
z_ind_real_poz = 3 + ((BMI - (M*(1+L*S*3)**(1/L)))/((M*(1+L*S*3)**(1/L)) - (M*(1+L*S*2)**(1/L))))
print(z_ind_real)
```
